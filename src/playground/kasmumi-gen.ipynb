```python
import urllib.request
from PIL import Image
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# download high res image
url = 'http://jinfo.jp/wp-content/uploads/2014/12/kasumi00.jpg'
filename = 'kasumi-high.jpg'
urllib.request.urlretrieve(url, filename)
```

```python
hr_image = cv2.imread('./{}'.format(filename))
```

```python
lr_image = mosaic(hr_image, 0.01)
```

```python
def mosaic(img, alpha):
    w = img.shape[1]
    h = img.shape[0]
    
    img = cv2.resize(img,(int(w*alpha), int(h*alpha)))
    img = cv2.resize(img,(w, h), interpolation=cv2.INTER_NEAREST)

    return img

def imshow(img):
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.show()
    
```

```python
imshow(hr_image)
imshow(lr_image)
```

```python
cv2.imwrite("kasumi-low.jpg", lr_image)
```
